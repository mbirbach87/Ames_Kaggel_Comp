```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

import statsmodels.api as sm

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
houses = pd.read_csv('trains.csv')
holdout_set = pd.read_csv('test.csv')
```

```python
houses.head()
```

```python
features = ['Overall Qual', 'Year Built']
x= houses [features]
y = houses ['SalePrice']
```

```python
# Now, let's build our GLM
indep_vars = ['Overall Qual', 'Year Built']
X = sm.add_constant(houses[indep_vars])
y = houses.SalePrice

glm_logit = sm.GLM(y, 
                   X,
                   sm.families.Binomial(sm.families.links.logit))
results_logit = glm_logit.fit()
```
